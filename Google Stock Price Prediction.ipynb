```python
# Loading neccessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Laoding the dataset

data_train= pd.read_csv(r'C:\Users\user\Desktop\Springboard\Curriculum Projects\Stock Price Prediction- Keras and RNN\Google_Stock_Price_Train.csv')
```

```python
# Initializing the training dataset
train= data_train.iloc[:, 1:6].values
train
```

```python

scaler= MinMaxScaler(feature_range= (0,6))
train_scaled= scaler.fit_transform(train)
```

```python
X_train=[]
y_train=[]

for i in range(60,1258):
    X_train.append(train_scaled[i-60:i,0])
    y_train.append(train_scaled[i,0])
X_train, y_train= np.array(X_train), np.array(y_train)

X_train= np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout

model= Sequential()
model.add(LSTM(units= 50, return_sequences= True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences= True))
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences= True))
model.add(Dropout(0.2))

model.add(LSTM(units=50))
model.add(Dropout(0.2))

model.add(Dense(units=1))

model.compile(optimizer= 'adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
data_test= pd.read_csv(r'C:\Users\user\Desktop\Springboard\Curriculum Projects\Stock Price Prediction- Keras and RNN\Google_Stock_Price_Test.csv')
test= data_test.iloc[:, 1:6].values
```

```python
data= pd.concat((data_train['Open'],
                 data_train['High'],
                 data_train['Low'],
                 data_train['Close'],
                 data_train['Volume'],
                 data_test['Open'],
                 data_test['High'],
                 data_test['Low'],
                 data_test['Close'],
                 data_test['Volume']), axis=0)
inputs= data[len(data)- len(data_test)- 60:].values
inputs= inputs.reshape(-1, 1)
inputs= scaler.transform(inputs)

X_test= []
for i in range(60,80):
    X_test.append(inputs[i-60:i, 0])
X_test= np.array(X_test)
X_test= np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
pred_stock_price= model.predict(X_test)
pred_stock_price= scaler.inverse_transform(pred_stock_price)

plt.plot(test, color= 'red', label= "Real Google Stock Price")
plt.plot(pred_stock_price, color= 'blue', label='Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```
